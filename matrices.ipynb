```python
sys=np.hstack((mat,x.reshape((2,1))))
print(sys)
```

```python

def plot_lines(sys):
    # sys is a 2×3 matrix like [[a1, b1, c1], [a2, b2, c2]]
    x = np.linspace(-10, 10, 400)

    for row in sys:
        a, b, c = row
        # Equation: a*x + b*y = c → y = (c - a*x)/b
        y = (c - a * x) / b
        plt.plot(x, y, label=f'{a}x + {b}y = {c}')

    # Solve for intersection
    A = sys[:, :2]
    B = sys[:, 2]
    intersection = np.linalg.solve(A, B)
    plt.scatter(*intersection, color='red', zorder=5, label=f'Intersection {tuple(np.round(intersection,2))}')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear System Lines')
    plt.axhline(0, color='black', lw=0.8)
    plt.axvline(0, color='black', lw=0.8)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_lines(sys)
```

### 1.1 - System of Linear Equations

Here is a **system of linear equations** (or **linear system**) with three equations and three unknown variables:


$$\begin{cases} 
4x_1-3x_2+x_3=-10, \\ 2x_1+x_2+3x_3=0, \\ -x_1+2x_2-5x_3=17, \end{cases}\tag{1}$$

**To solve** this system of linear equations means to find such values of the variables $x_1$, $x_2$, $x_3$, that all of its equations are simultaneously satisfied.

```python
mat3=np.array([[4,-3,1],[2,1,3],[-1,2,-5]])
vec=np.array([-10,0,17])
ans=np.linalg.solve(mat3,vec)
print(ans)
```
